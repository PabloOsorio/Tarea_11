```python
class BinaryNode:
    def __init__( self , value , left=None , rigth= None ):
        self.data = value
        self.left = left
        self.rigth = rigth
 
class BinarySearchTree:
    def __init__( self ):
        self.root = None
    
    def insert( self , value ):
        if self.root == None:
            self.root = BinaryNode( value )
        else:
            self.__insert_nodo( self.root , value )
        
    def __insert_nodo(self , nodo , value ):
        if nodo.data == value: 
            pass
        elif value < nodo.data:# ir a la izq
            if nodo.left == None:
                nodo.left = BinaryNode( value )
            else:
                self.__insert_nodo( nodo.left , value )
        else:  # ir a la derecha
            if nodo.rigth == None:
                nodo.rigth = BinaryNode( value )
            else:
                self.__insert_nodo( nodo.rigth , value )
    
    def transversal( self , formato = "posorden"):
        if formato == "posorden":
            self.recorrido_pos( self.root )
        elif formato == "preorden":
            self.recorrido_pre(self.root )
        else: # inorden
            self.recorrido_in( self.root )
        print()
 
    def  recorrido_pos( self , nodo ):
        if nodo != None:
            self.recorrido_pos( nodo.left )
            self.recorrido_pos( nodo.rigth )
            print( str(nodo.data) , end=" , " )
    
    def  recorrido_pre( self , nodo ):
        if nodo != None:
            print( str(nodo.data) , end=" , " )
            self.recorrido_pre( nodo.left )
            self.recorrido_pre( nodo.rigth )
        else:
            return
    
    def  recorrido_in( self , nodo ):
        if nodo != None:
            self.recorrido_in( nodo.left )
            print( str (nodo.data) , end=" , " )
            self.recorrido_in( nodo.rigth ) 
 
    def search( self , value ):
        if self.root == None:
            print("El árbol está vacío")
            return None
        else:
            return self.__search( self.root , value )
 
    def __search( self , nodo , value ):
        if nodo == None:
            print("El dato no existe en el árbol") 
            return None
        elif nodo.data == value:
            print("Encontrado", nodo.data)
            return nodo
        elif value < nodo.data:
            return self.__search( nodo.left , value  )
        else:
            return self.__search( nodo.rigth , value  )
        
    def borrar( self , value ):
        if self.root == None:
            print("El árbol está vacío")
            return None
        else:
            return self.__borrar( self.root , value )

    def __borrar( self , nodo , value ):
        if nodo == None:
            print("el dato no existe en el arbol") 
            return None

        elif nodo.data == value:
            print("Encontrado y eliminado", nodo.data)
            #Encontrado primero despues lo borramos
            if nodo.left != None and nodo.rigth == None:
                print("Hijo izq")
                nodo.data = None
                nodo.data = nodo.left.data
            elif nodo.left == None and nodo.rigth != None:
                print("Hijo derecha")
                nodo.data = None
                nodo.data = nodo.rigth.data
            else:
                print("Elimine una hoja")
                nodo.data = None
                nodo = nodo.data 
                return

        elif value < nodo.data:
            return self.__borrar( nodo.left , value  )
        else:
            return self.__borrar( nodo.rigth , value ) 
```
